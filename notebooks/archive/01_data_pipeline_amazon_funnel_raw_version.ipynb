```python
categories = ["board game", "coffee beans", "wireless headphones"]
```

```python
all_products = []

for category in categories:
    print(f"\nFetching: {category}")
    
    for page in range(1, 3):  # 2 pages per category is usually enough for free tier
        print(f"Page: {page}")
        
        params = {
            "keyword": category,
            "page": str(page),
            "domainCode": "com",
            "excludeSponsored": "false",
            "sortBy": "relevanceblender",
            "withCache": "true"
        }

        response = requests.get(url, headers=headers, params=params)
        
        print("Status:", response.status_code)
        
        data = response.json()
        
        products = data.get("searchProductDetails", [])
        
        print("Products found:", len(products))
        
        for p in products:
            p["category"] = category
            all_products.append(p)

        time.sleep(1)
```

```python
def clean_sales_volume(x):
    if pd.isna(x):
        return np.nan
    
    x = str(x).upper()
    
    num = ""
    for char in x:
        if char.isdigit():
            num += char
        elif char == ".":
            num += char
        elif num:
            break 
    
    if num == "":
        return np.nan
    
    value = float(num)
    
   
    if "K" in x:
        value = value * 1000
    
    return value
    
```
