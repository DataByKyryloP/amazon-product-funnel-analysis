```python
df_final["has_reviews"] = (df_final["countReview"] > 0).astype(int)
```

```python
df_final["is_discounted"] = (df_final["price"] < df_final["retailPrice"]).astype(int)
```

```python
import numpy as np

df_final["retailPrice"] = df_final["retailPrice"].replace(0, np.nan)
```

```python
df_final["is_discounted"] = (
    (df_final["price"] < df_final["retailPrice"])
).astype(float)
```

```python
df_final[["price", "retailPrice", "is_discounted"]].sample(10, random_state=42)
```

```python
df_final["is_discounted"] = np.where(
    df_final["retailPrice"].notna() & df_final["price"].notna(),
    (df_final["price"] < df_final["retailPrice"]).astype(int),
    np.nan
)
```

```python
df_final[["price", "retailPrice", "is_discounted"]].sample(10, random_state=42)
```

```python
df_final.to_csv("amazon_products_clean_v3final.csv", index=False)
```
